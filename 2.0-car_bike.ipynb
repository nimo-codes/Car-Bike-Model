```python
from keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
import keras
```

This car-bike is a CNN model which is based on a sequential model and uses keras and tensorflow.
it is successfully able to predict with the accuracy of 94.18 %.

lets start off with importing all the required libraries.

```python
train_dir = "./car-bike/Train_set/"
test_dir = "./car-bike/Test_set/"
```

we divided the dataset into 2 folders which inlcudes training and testing folders and then each of the folders have two folders in them which has bike and car pictures in them.

lets define a variable to give the location of the folders.

```python
train_datagen = ImageDataGenerator(
        rotation_range=10,
        width_shift_range=0.2, 
        height_shift_range=0.2, 
        zoom_range=0.2, 
        horizontal_flip=True, 
        brightness_range=[0.2,1.2]) 
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150,150),
    batch_size=128,
    class_mode='binary'
)
```

now we use image data generator from the tensorflow library to help create more random images based on the particulars we give they inlcude brightness range - which will basically increase or decrease brightness of the images varying in a defined range, zoom range will basically zoom the images, flipping will flip the images horizontally or verticaly whichever is specified.

then we specify the target size for the images which is 150x150 so that all the images have same format and we use class mode as binary as we are differentiating between two types of images.

```python
test_datagen = ImageDataGenerator(
        rotation_range=10,
        width_shift_range=0.2, 
        height_shift_range=0.2, 
        zoom_range=0.2, 
        horizontal_flip=True, 
        brightness_range=[0.2,1.2]) 
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150,150),
    batch_size=128,
    class_mode='binary'
)
```

we do the same things as previously written above for the test images also

```python
model = keras.models.Sequential([
    keras.layers.Conv2D(filters=32,kernel_size=3,strides=(1,1),padding="same",activation='relu',input_shape=[150,150,3]),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D((2,2)),
    keras.layers.Dropout(0.3),
    keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu'),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D((2,2)),
    keras.layers.Dropout(0.3),
    keras.layers.Conv2D(filters=128,kernel_size=3,activation='relu'),
    keras.layers.BatchNormalization(),
    keras.layers.Flatten(),
    keras.layers.Dense(units=128,activation ='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(units=1,activation ='sigmoid')

])
from tensorflow.keras.optimizers import RMSprop,SGD
model.compile(loss = 'binary_crossentropy',
                optimizer = SGD(learning_rate=0.005),
                metrics=['accuracy'])
```

Now we start off making our model.

at first we will make a sequential model which goes step by step from the above.

secondly we use conv2d from keras and give it the basic attributes such as input shape of the images, its 150x150 and 3 as we are giving its rgb values. kernel size is basically a matrix which will go accross the pixels of the image, padding is like the pixels which are left have to be calculated we are using "same" , it will put the leftover values with 0 and calculate, filters are basically how many times the kernels gonna move across the image

then we use batch normalization which will rescale and recenter the artificial neural networks and make the model stable.

then we use dropout to drop some of the neurons in the network as it helps reducing overfitting.

the activation function  we use is reLu which gives value as 0 for negative and 0 values and 1 otherwise.


```python
model.summary()
```

```python
from keras.callbacks import ModelCheckpoint
model_path = "/Users/jarvis/pymycod/tensorflow_AI/trained_models/lr_model_carsvsbike.h5"
checkpoint = ModelCheckpoint(model_path,monitor="val_accuracy",verbose=1,save_best_only=True,mode=max)
callbacks_list=[checkpoint]
```

we use callback function of the keras to make callbacks and check the accuracy so we can save the highest model to our system , verbose means the amount of info we need for a particular epoch

```python
history = model.fit(train_generator,epochs=100,verbose=2,batch_size=128,validation_data=test_generator,callbacks=callbacks_list)
```

now at last we fit the model with the train and generalized images and also give the test set for validating.

we run it for 100 epochs and the batch size is 128.

Accuracy of the model is 94.18%, this accuracy is based on the validation set which is test dataset
